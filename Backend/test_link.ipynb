```python
from transformers import pipeline
from huggingface_hub import notebook_login
import pandas as pd
```

```python
path_data = r'C:\IA\Test_link\test_data.csv'
df = pd.read_csv(path_data)
df.text.map(len)
df.text = df.text.astype(str)
```

```python
df.text.to_list()
```

```python
notebook_login()
```

```python
classifier = pipeline("text-classification", model="JeisonJA/distilbert-pid-smartlink-V2", tokenizer="distilbert-base-cased")
```

```python
prediction = classifier(df.text.to_list())
```

```python
#print(prediction)
```

```python
df['label_predict'] = [predict['label'] for predict in prediction]
df.head(1000)
```

```python
df['label_predict'] = df['label_predict'].str.replace('LABEL_', '') 
```

```python
df['label'] = df['label'].astype(str)
```

```python
print(df)
```

```python
valores_consistentes = (df['label'] == 0).sum()
valores_inconsistentes = (df['label'] != 0).sum()
print('Elementos consistentes:', valores_consistentes)
print('Elementos Inconsistentes:', valores_inconsistentes)
```

```python
resume = df['label'].value_counts()
print(resume)
```

```python
coincidente = (df['label'] == df['label_predict']).sum()
diferentes = (df['label'] != df['label_predict']).sum()
print('Resumen \n', 'Predicciones correctas:\n', coincidente, '----->', (coincidente/84)*100, '%', '\n', '\n',  'Predicciones incorrectas:\n', diferentes, '----->', (diferentes/84)*100, '%')
```
